```python
from utils import utils
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
main = pd.read_csv(r'./data/train_dm300.csv')
```

```python
solvents = ["MeOH", "EtOH", "IPA", "Tol", "Hept", "Acet", "MeCN", "MTBE", "MeTHF", "DMF", "EtOAc"]
solvent_names = {"Methanol":"Methanol", 
                "Ethanol":"Ethanol", 
                "2-Propanol":"2-Propanol", 
                "Toluene":"Toluene", 
                "Heptane":"Heptane", 
                "Acetone":"Acetone", 
                "Acetonitrile":"Acetonitrile", 
                "Methyl tert butyl ether":"Methyl\ntert-butyl ether", 
                "2-Methly tetrahydrofurane":"2-Methyl\ntetrahydrofurane",
                "Dimethyl formamide":"Dimethyl\nformamide", 
                "Etyl acetate":"Ethyl\nacetate" }

solvent_dict = {}
for solvent, name in zip(solvents, solvent_names.keys()):
    solvent_dict[solvent] = main[main['solvent_name']==name]
```

# Donut chart

```python
sizes = [np.count_nonzero(solvent['dm300']) for solvent in solvent_dict.values()]
colors = [
        "#8dd3c7",
        "#ffffb3",
        "#bebada",
        "#fb8072",
        "#80b1d3",
        "#fdb462",
        "#b3de69",
        "#fccde5",
        "#d9d9d9",
        "#bc80bd",
        "#ccebc5"]
```

```python

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(aspect="equal"), tight_layout=True)

my_circle = plt.Circle((0,0), 0.3, color='white')

data = sizes
ingredients = solvent_names.values()


def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d})".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="black"), colors=colors)

ax.legend(wedges, ingredients,
          title="Solvents",
          loc="best",
          bbox_to_anchor=(1, 0, 0.5, 1), ncol=11)

plt.setp(autotexts, size=8)#, weight="bold")
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.savefig(r'./figures/donut-chart-of-total.png')
plt.savefig(r'./figures/donut-chart-of-total.svg')
```

# Bar chart on avg rejection

```python
avg_rejections = [round(np.average(solvent['dm300']), 3) for solvent in solvent_dict.values()]
```

```python
fig, ax = plt.subplots(figsize=(5,5), tight_layout=True)
bars = ax.bar(solvent_names.values(), avg_rejections, color=colors)
plt.ylim(-0.1, 1)
plt.ylabel("Average rejection (-)")
y_pos = range(len(solvent_names.values()))
plt.xticks(y_pos, solvent_names.values(), rotation=90)
ax.bar_label(bars)
plt.savefig(r'./figures/bar-chart-average-rejection.png')
plt.savefig(r'./figures/bar-chart-average-rejection.svg')
plt.show()
```

# Bar chart on error

```python
avg_error = [round(np.average(solvent['dm300_error']), 3) for solvent in solvent_dict.values()]
```

```python
fig, ax = plt.subplots(figsize=(5,5), tight_layout=True)
bars = ax.bar(solvent_names.values(), avg_error, color=colors)
plt.ylim(0, 0.15)
plt.ylabel("Average error (-)")
y_pos = range(len(solvent_names.values()))
plt.xticks(y_pos, solvent_names.values(), rotation=90)
ax.bar_label(bars)
plt.savefig(r'./figures/bar-chart-average-error.png')
plt.savefig(r'./figures/bar-chart-average-error.svg')
plt.show()
```
